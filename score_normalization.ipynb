## Would using a saturation function produce the desired behavior?

```python
import pandas as pd
import math

plt.rcParams['figure.figsize'] = [20, 6]
figs, axes = plt.subplots(ncols=2, nrows=1)

def sigmoid(x):
    result = 1.0 / (1.0 + math.exp(-1/30 * x))
    return round(result, 6)

def sigmoid_vec(v):
    return [sigmoid(x) for x in v]

plot("baseline", axes[0], {
    "query_expansion": query_expansion_scores,
    "join_tables": join_scores,
    "explain": explain_scores,
    "delete_by_query": delete_by_query_scores
})

plot("sigmoid", axes[1], {
    "query_expansion": sigmoid_vec(query_expansion_scores),
    "join_tables": sigmoid_vec(join_scores),
    "explain": sigmoid_vec(explain_scores),
    "delete_by_query": sigmoid_vec(delete_by_query_scores)
})

plt.show()
```
